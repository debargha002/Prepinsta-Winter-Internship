### 3. Outliers: Identify potential outliers in numerical indicators (e.g., Age, Salary, Rating). Should outliers be removed or adjusted?

Using the *IQR* method, I have computed outliers for numerical columns like `rating` and `age`. As can be seen from the provided data, the rating column has some missing data and some data is formatted incorrectly. Therefore, before looking for outliers, I corrected the incorrect data, handled missing values by taking the mode of the other values, and handled the "-1" by taking the absolute of the column, which led to the conversion of any negative value into a positive one (since a rating cannot be negative).

```python
df2=df["Rating"].mode()[0]
df["Rating"]=df["Rating"].fillna(df2)
df["Rating"]=df["Rating"].abs()

Out=['Age','Rating']
for i in Out:

    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[i] < lower_bound) | (df[i] > upper_bound)]

    print(f"Potential outliers in {i}:")
    print(outliers)
```

### 4. Salary Formatting: Examine the format of the Salary column. Does it require any formatting or standardization for consistent analysis?

I changed the format of the salary to be more readable and consistent for the analysis by removing the $ and replacing the k with 000.

```python
df["Salary"]=df["Salary"].replace(['\$']," ",regex=True)
df["Salary"]=df["Salary"].replace('k',"000",regex=True)
df
```

### 5. Location Standardization: Check the consistency of location entries. Do they need standardization, and how can this be achieved?

I've noticed that every nation is represented using the same format, which consists of the name of the nation plus its two initials. I had removed everything after the comma to leave only the country name for better standardization.

```python
df['Location'] = df['Location'].replace("Australia Aus", "Australia, Au")
split_locations = df['Location'].str.split(',', n=1, expand=True)
df['Location'] = split_locations[0].str.strip()
df
```

### 7. Easy Apply Indicator: Analyze the Easy Apply column. Does it contain boolean values or need transformation for better analysis?

```python
print("\nFirst Approach : \n") # By replacing -1 by FALSE to have constant boolean values
df3=df["Easy Apply"].replace("-1","FALSE")
print(df3)


print("\nSecond Approach : \n") # By replacing -1 by 0 and TRUE by 1 to have constant Numerical values 
df4=df["Easy Apply"].replace("TRUE","1")
df5=df4.replace("-1","0")
print(df5)

df['Easy Apply']=df5 # I changed the dataset's Easy Apply column to a numerical format (True=1, False=0) in order to conduct additional research.
```
